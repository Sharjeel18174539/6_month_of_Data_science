Removing Null values
2 step either remove the null value or fill the row 

Step-1 Remove the row which have null values

```python
# df.dropna(inplace=True)
print(df.info())
```

We can fill the value using mean median mode

```python
print(df['Item_Weight'])
```

```python
# As we know Item_weight is float type it's numeric so we use mean or median or mode any thing to fill the value lets fill with mean
x_mean=df['Item_Weight'].mean()
df.fillna({'Item_Weight':x_mean,}, inplace=True)
print(df.isnull().sum())
```

```python
print(df['Outlet_Size'])
```

```python
y=df['Outlet_Size'].mode()
print(y)
```

```python
y=df['Outlet_Size'].mode()[0]
df.fillna({'Outlet_Size':y}, inplace=True)
print(df.isnull().sum())
```

```python
print(df.duplicated())
```

```python
# calculate the averages of the item
avg=df['Item_Weight'].mean()
print(avg)
```
